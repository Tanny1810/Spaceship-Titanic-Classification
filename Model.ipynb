### One-hot Encoding

```python
X_encoded = pd.get_dummies(X, columns = ['HomePlanet', 'Destination'])
X_encoded.head()
```

```python
test_df = pd.get_dummies(test_df, columns = ['HomePlanet', 'Destination'])
test_df.head()
```

```python
X_encoded = X_encoded.drop('Transported', axis=1)
X_encoded.head()
```

```python
y = X['Transported'].copy()
y.head()
```

```python
X_encoded.dtypes
```

```python
X_encoded = X_encoded.drop('PassengerId', axis=1)
X_encoded.head()
```

```python
X_encoded.corr()
```

## Model

```python
sum(y)/len(y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_encoded, y, random_state=42, stratify=y)
```

```python
sum(Y_train)/len(Y_train)
```

```python
sum(Y_test)/len(Y_test)
```

```python
xgb_class = xgb.XGBClassifier(objective='binary:logistic', seed=42, use_label_encoder=False)
xgb_class
```

```python
xgb_model = xgb_class.fit(X_train, Y_train, early_stopping_rounds=10, eval_metric='aucpr', eval_set=[(X_test, Y_test)])
```

```python
predictions = xgb_class.predict(X_test)
cm = confusion_matrix(Y_test, predictions)
cm_disp = ConfusionMatrixDisplay(confusion_matrix= cm, display_labels=["Transported", "Not_Transported"])
cm_disp.plot()
plt.show()
```

```python
xgb.plot_importance(xgb_model)
```

## Parameter Optimization

```python
param_grid = {
    'max_depth':[5,6,7],
    'learning_rate':[0.01, 0.05, 0.005],
    'gamma':[2.5,3.0,3.5],
    'reg_lambda':[1.0,3.0],
    'scale_pos_weight':[1,3,5]}
```

```python
optimal_param = GridSearchCV(estimator=xgb.XGBClassifier(objective='binary:logistic',
                                                        missing = 1,
                                                        seed = 42,
                                                        subsample = 0.5,
                                                        colsample_bytree = 0.5), 
                            param_grid= param_grid,
                            scoring='roc_auc',
                            n_jobs=10,
                            cv=3)
```

```python
optimal_param.fit(X_train, Y_train, early_stopping_rounds = 10, eval_metric = 'auc', eval_set = [(X_test, Y_test)])
```

```python
print(optimal_param.best_params_)
```

```python
xgb_class_opti = xgb.XGBClassifier(seed = 42, 
                              objective = 'binary:logistic', 
                              gamma = 2.5,
                              learning_rate = 0.05,
                              max_depth = 6,
                              reg_lambda = 1.0,
                              scale_pos_weight = 1,
                              subsample = 0.9, 
                              colsample_bytree = 0.5)
```

```python
xgb_class_opti.fit(X_train, Y_train, early_stopping_rounds=10, eval_metric='aucpr', eval_set=[(X_test, Y_test)])
```

```python
predictions_opti = xgb_class_opti.predict(X_test)
cm = confusion_matrix(Y_test, predictions_opti)
cm_disp = ConfusionMatrixDisplay(confusion_matrix= cm, display_labels=["Transported", "Not_Transported"])
cm_disp.plot()
plt.show()
```
